```python
import numpy
import matplotlib.pyplot as mt
```

```python
list2 = numpy.array([5,10])
list3 = numpy.array([3,4])

mt.plot(list2,list3)
mt.show()
```

```python
x = numpy.random.normal(170,10,250)
mt.hist(x)  
mt.show()
```

```python
y = numpy.array([20,30,50])
label = ["apple","banana","grapes"]

mt.pie(y, labels=label)
mt.show()
```
